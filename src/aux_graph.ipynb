```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
models = ["LightGCN", "LF-MMLGCN", "LF-MMLGCN (alpha)", "LF-MMLGCN (unfrozen)", "LF-MMLGCN (normalized)"]
# Metric@10
precision = [0.0488, 0.0618, 0.0605, 0.0563, 0.0577]
recall = [0.0759, 0.0985, 0.0962, 0.0901, 0.0923]
ndcg = [0.0763, 0.0961, 0.0931, 0.0890, 0.0902]
```

```python
# Grouped bar chart (group by metric)

model_scores = {
    "LightGCN": [precision[0], recall[0], ndcg[0]],
    "LF-MMLGCN": [precision[1], recall[1], ndcg[1]],
    "LF-MMLGCN (alpha)": [precision[2], recall[2], ndcg[2]],
    "LF-MMLGCN (unfrozen)": [precision[3], recall[3], ndcg[3]],
    "LF-MMLGCN (normalized)": [precision[4], recall[4], ndcg[4]]
}

N_MODELS = len(model_scores)

metrics = ["Precision", "Recall", "NDCG"]
all_scores = [precision, recall, ndcg]

x = np.arange(len(metrics))
width = 0.8 / N_MODELS  # N models, distribute 80% of space

fig, ax = plt.subplots(figsize=(max(8, len(metrics) * 2), 6))

# Calculate bar positions
bar_positions = []
start_offset = -(N_MODELS - 1) * width / 2
for i in range(N_MODELS):
    bar_positions.append(x + start_offset + i * width)

# Create bars for each model
for i, model_name in enumerate(model_scores.keys()):
    model_scores = [precision[i], recall[i], ndcg[i]]
    ax.bar(bar_positions[i], model_scores, width, label=model_name, alpha=0.8)

ax.set_xlabel("Metrics")
ax.set_ylabel("Score")
ax.set_title("Model performance comparison (grouped by metric)\nDbBook, k=10")
ax.set_xticks(x)
ax.set_xticklabels(metrics)
#ax.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Grouped bar chart (group by metric)

x = np.arange(len(models))
width = 0.25

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(x - width, precision, width, label="Precision", alpha=0.8)
ax.bar(x, recall, width, label="Recall", alpha=0.8)
ax.bar(x + width, ndcg, width, label="NDCG", alpha=0.8)

ax.set_xlabel("Models")
ax.set_ylabel("Score")
ax.set_title("Model performance comparison (grouped by model)\nDbBook, k=10")
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

plt.tight_layout()
plt.show()
```
